```python
plt.figure(figsize=(12, 6))
sns.histplot(dataset['value [USD]'], bins=30,color='blue')
plt.axvline(dataset["value [USD]"].mean(), color = "red", linestyle="--", label="Average Purchase in USD")
plt.title("Distribution of Purchase Value")
plt.xlabel("Value of Purchase in USD")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pay_count = dataset["payment_method"].value_counts()
sns.barplot(x=pay_count.index, y=pay_count.values, palette = "viridis")
plt.title("Number of Transactions by Payment Method")
plt.xlabel("Payment Methods")
plt.ylabel ("count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
product_category = dataset["product_category"].value_counts()
sns.barplot(x=product_category.index, y=product_category.values, palette = "viridis")
plt.title("Number of Product sales by Categories")
plt.xlabel("Product Category")
plt.ylabel ("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='product_category', y='value [USD]', data=dataset)
plt.title("Product category by Sales Value")
plt.xlabel("Product Category")
plt.ylabel ("Sales value in USD")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='payment_method', y='value [USD]', data=dataset, palette="pastel")
plt.title("Transaction Value by Payment Method")
plt.show()
```
