```python
#Clean data to remove empty rows and convert age and salary to integers
data.isnull().sum()
data
data_cleaned = data.dropna()
data_cleaned['Age'] = data_cleaned['Age'].astype(int)
data_cleaned['Salary'] = data_cleaned['Salary'].astype(int)

data_cleaned
```

```python
data_cleaned.describe()
```

```python
data_cleaned.info()
```

```python
#Data Filtering
Marketing_Department = data_cleaned[data_cleaned["Department"]=="Marketing"]
Marketing_Department
```

```python
#Data Filtering
Senior_Staff = data_cleaned[data_cleaned["Age"]>=40]
Senior_Staff
```

```python
#Data Filtering
Low_Salary = data_cleaned[data_cleaned["Salary"]<=60000]
Low_Salary
```

```python
#Data Filtering Joining two Columns together
Finance_Salary = data_cleaned[(data_cleaned["Department"]=="Finance") & (data_cleaned["Salary"]>=50000)]
Finance_Salary
```

```python
#DATA SORTING
sorting_age = data_cleaned.sort_values(by='Age')
sorting_age
```

```python
#DATA SORTING
sorting_salary = data_cleaned.sort_values(by='Salary')
sorting_salary
```

```python
# Data Sorting
sorting_data = data_cleaned.sort_values(by=['Age', 'Salary'], ascending=[True, False])
sorting_data
```
